```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Student Info.csv")
df.head()
```

```python
categorical_col = list(df.select_dtypes(include=['object']).columns)
```

```python
one_hot_encoded_data = pd.get_dummies(df, columns = categorical_col) 
for i in one_hot_encoded_data.select_dtypes(include=['bool']).columns:
    one_hot_encoded_data[i] = one_hot_encoded_data[i].astype(int)
```

```python
one_hot_encoded_data
```

```python
X = one_hot_encoded_data.drop(columns=["G3"], axis=1)
y = one_hot_encoded_data['G3']
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split
import numpy as np

def train(model, X, y):
    # train the model
    x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42)
    model.fit(x_train, y_train)
    
    # predict the training set
    pred = model.predict(x_test)
    
    # perform cross-validation
    cv_score = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
    cv_score = np.abs(np.mean(cv_score))
    
    # calculate R2 score
    r2 = r2_score(y_test, pred)
    
    print("Model Report")
    print("MSE:", mean_squared_error(y_test, pred))
    print('CV Score:', cv_score)
    print('R2 Score:', r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
train(model, X, y)
```

```python
import lightgbm as lgb
model_1 = lgb.LGBMRegressor(objective='regression', random_state=42)
train(model_1, X, y)
```
